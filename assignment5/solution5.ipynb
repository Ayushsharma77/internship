```python
import requests

def weather_details(city, api_key):
    url = "https://api.openweathermap.org/data/2.5/weather?"
    final_url = f"{url}q={city}&appid={api_key}&units=metric"

    try:
        response = requests.get(final_url)
        response.raise_for_status()
        data = response.json()

        # basic info (from notes)
        print(f"\nWeather in {city}:")
        print(f"Temperature   : {data['main']['temp']} C")
        print(f"Feels Like    : {data['main']['feels_like']} C")
        print(f"Description   : {data['weather'][0]['description'].capitalize()}")

        # extra data added by studying the API response structure
        print(f"Min Temp      : {data['main']['temp_min']} C")
        print(f"Max Temp      : {data['main']['temp_max']} C")
        print(f"Humidity      : {data['main']['humidity']} %")
        print(f"Pressure      : {data['main']['pressure']} hPa")
        print(f"Wind Speed    : {data['wind']['speed']} m/s")
        print(f"Wind Direction: {data['wind']['deg']} degrees")
        print(f"Cloud Cover   : {data['clouds']['all']} %")
        print(f"Visibility    : {data.get('visibility', 'N/A')} meters")
        print(f"Country       : {data['sys']['country']}")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching weather data: {e}")

api_key = "b8aeb299f93d443bd5cbeecdb72fc733"
city = input("Enter city name: ")
weather_details(city, api_key)
```

```python
import random

options = ["rock", "paper", "scissor"]

while True:
    ch = int(input('''
--- ROCK PAPER SCISSOR ---
1. Play
2. Exit
Enter choice: '''))

    if ch == 1:
        uscore = 0
        cscore = 0

        for round in range(1, 6):
            print(f"\n--- Round {round} ---")
            u = int(input('''1. Rock
2. Paper
3. Scissor
Your choice: '''))

            if u == 1:
                uch = "rock"
            elif u == 2:
                uch = "paper"
            elif u == 3:
                uch = "scissor"
            else:
                print("Invalid input, skipping round")
                continue

            cch = random.choice(options)
            print(f"You chose: {uch}  |  Computer chose: {cch}")

            if uch == cch:
                print("Round Draw!")
                uscore += 1
                cscore += 1
            elif ((uch == "rock"    and cch == "scissor") or
                  (uch == "paper"   and cch == "rock")    or
                  (uch == "scissor" and cch == "paper")):
                print("You win this round!")
                uscore += 2
            else:
                print("Computer wins this round!")
                cscore += 2

        print(f"\n=== Final Score ===")
        print(f"You: {uscore}  |  Computer: {cscore}")

        if uscore == cscore:
            print("Game Drawn!")
        elif uscore > cscore:
            print("You Won the Game!")
        else:
            print("Computer Won the Game!")

    else:
        print("Thanks for playing!")
        break
```

```python
import requests

def get_jokes():
    url = "https://v2.jokeapi.dev/joke/Any?amount=5"

    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()

        print("\n--- 5 Random Jokes ---\n")
        for i, joke in enumerate(data['jokes'], 1):
            print(f"Joke {i} | Category: {joke['category']}")
            if joke['type'] == 'single':
                print(f"  {joke['joke']}")
            else:
                print(f"  Q: {joke['setup']}")
                print(f"  A: {joke['delivery']}")
            print()

    except requests.exceptions.RequestException as e:
        print(f"Error: {e}")

get_jokes()
```
