```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2], 'B': [4, 5]})
for index, row in df.iterrows():
    print(index, row['A'], row['B'])
for row in df.itertuples():
    print(row.Index, row.A, row.B)
```

```python
import pandas as pd
df = pd.DataFrame({'A': [1, 3], 'B': [4, 6]}, index=[0, 2])
new_row = pd.DataFrame({'A': [2], 'B': [5]}, index=[1])
df_inserted = pd.concat([df.iloc[:1], new_row], axis=0)
df_inserted = pd.concat([df_inserted, df.iloc[1:]], axis=0)
print(df_inserted)
```
