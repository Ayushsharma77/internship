```python
import pandas as pd
date_strings = pd.Series([
    '2026-01-10', '2026-02-15', '2026-03-20',
    '2026-04-25', '2026-05-30', '2026-06-05'
])

timeseries = pd.to_datetime(date_strings)

values = pd.Series([100, 150, 200, 130, 175, 220], index=timeseries)
print("Time-indexed Series:")
print(values)

print("\nAccessing by year 2026:")
print(values['2026'])

print("\nMonth from each date:")
print(timeseries.dt.month)

print("\nDay name for each date:")
print(timeseries.dt.day_name())
```

```python
df1 = pd.DataFrame({
    'ID':   [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eva'],
    'Dept': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

df2 = pd.DataFrame({
    'ID':     [2, 3, 4, 6, 7],
    'Salary': [55000, 62000, 48000, 70000, 53000],
    'City':   ['Delhi', 'Mumbai', 'Pune', 'Chennai', 'Kolkata']
})

# Inner Merge
inner = pd.merge(df1, df2, on='ID', how='inner')
print("\nInner Merge:")
print(inner)

# Left Join
left_join = pd.merge(df1, df2, on='ID', how='left')
print("\nLeft Join:")
print(left_join)

# Right Join
right_join = pd.merge(df1, df2, on='ID', how='right')
print("\nRight Join:")
print(right_join)

# Index-Based Join
df1_idx = df1.set_index('ID')
df2_idx = df2.set_index('ID')
index_join = df1_idx.join(df2_idx, how='left')
print("\nIndex-Based Join:")
print(index_join)
```

```python
dept_A = pd.DataFrame({
    'EmpID': [101, 102, 103],
    'Name':  ['Arjun', 'Bhavna', 'Chirag'],
    'Dept':  ['IT', 'IT', 'IT']
})

dept_B = pd.DataFrame({
    'EmpID': [201, 202, 203],
    'Name':  ['Diya', 'Eshan', 'Fiona'],
    'Dept':  ['HR', 'HR', 'HR']
})

salaries = pd.DataFrame({
    'EmpID':  [101, 102, 103, 201, 202, 203],
    'Salary': [60000, 72000, 55000, 48000, 51000, 67000],
    'Grade':  ['B', 'A', 'C', 'C', 'C', 'B']
})

# Concatenate vertically
combined = pd.concat([dept_A, dept_B], ignore_index=True)

# Merge with the third dataframe
final = pd.merge(combined, salaries, on='EmpID', how='inner')
print("\nFinal Concatenated and Merged DataFrame:")
print(final)
```
