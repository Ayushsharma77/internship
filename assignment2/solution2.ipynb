```python
#Q1 solution
#dictionaries

user = {
    "name" : "sam",
    "age" : 21
}
print("dict : ", user)

user["city"] = "delhi"
user["age"] = 22

print("name is :", user["name"])
user.pop("city")
print("updated dict: ", user)

#tuples
nums = (0,1,1,2,3)
print("\n tuples : ", nums)

print("first item : ", nums[0])

print("count of 1: ", nums.count(1) )

print("index of 1 : ", nums.index(1))

#set
A = {1,2,3,3}
B = {3,4,5}
print("set A :", A)
A.add(6)
A.discard(1)
print("updated set A :", A)

print("A union B", A.union(B))
print("A intersection B", A.intersection(B))
print("A - B", A.difference(B))
```

```python
#Q3 solution
def calculate(x,y, choice) :
    if(choice == "add"):
        return x+y
    elif(choice == "substract"):
        return x-y
    elif(choice == "multiply"):
        return x*y
    elif(choice == "divide"):
        if(y != 0):
            return x/y
        else:
            return "cannot divide by 0!"
    else:
        return "Invalid operation"
    





x = float(input("enter first number :  "))
y = float(input("enter secopnd number :  "))
choice = input("Enter choice : add, substract, multiply, divide").lower().strip()
result = calculate(x,y,choice)
print("Result : ", result)
```

```python
orignal_num = int(input("Enter the number to check : "))
reverse_num = 0
temp = orignal_num

while temp>0:
    last_digit = temp%10
    reverse_num = reverse_num*10 + last_digit
    temp = temp//10

print("reverse num : ", reverse_num)

if(reverse_num == orignal_num):
    print(f"{orignal_num} is a palindorme")
else:
    print(f"{orignal_num} is not a palindorme")
```
