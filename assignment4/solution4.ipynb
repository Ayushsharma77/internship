```python
def function(num1,num2,num3):
    
    if(num1>num2 and num1>num3):
        return num1
    if(num2>num1 and num2>num3):
        return num2
    if(num3>num2 and num3>num1):
        return num3
    
max = function(90,40,30)
print(f"max among three number is {max}")
```

```python
def function(inputlist):
    multiply = 1
    for i in inputlist:
        multiply*=i
    print("multiplication of list element result :",multiply)



l = [10,20,20,10,50,9,17,66]
function(l)
```

```python
def function(num):
    fact = num
    if(fact == 1 or fact == 0):
        return 1
    if(num<0):
        return print("factorial is not defined for negative numbers")
    return fact*function(num-1)
print(function(5))

```

```python
def function(num):
    if(num>=0 and num<=100):
        print(f"{num} falls in range 0-100")
    elif(num>100 and num<=1000):
        print(f"{num} falls in range 101-1000")
    else:
        print(f"{num} is greater than 1000")
function(1000)
```

```python
def function(inputlist):
    outputlist = []
    for i in inputlist:
        if(i%2 == 0):
            outputlist.append(i)
    print(f"even numbers in list : {outputlist}")


l = [1,2,4,66,22,4,5,7,16,88,99,101]
function(l)
```

```python
def function(num):
    flag = 1
    for i in range(2,int(num/2)+1):

        if(num%i==0):
            flag = 0
            break

    if(flag == 0):
        print("number is not prime")
    else:
        print("number is prime")
    
function(17)
```

```python
def function(st):
    upper_count = 0
    lower_count = 0

    for char in st:
        if(char.isupper()):
            upper_count+=1
        elif(char.islower()):
            lower_count+=1
    
    print(f"uppercase count: {upper_count}\nlowercase count: {lower_count}")



st = "AyushSharma is my NAME"
function(st)
```
