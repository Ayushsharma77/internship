```python
import csv

# writing data to csv file - taking input from user
with open('address_book.csv', 'w') as file:
    writer = csv.writer(file)
    
    # writing header row first
    writer.writerow(['Name', 'Address', 'Mobile', 'Email'])
    
    # taking 3 entries from user
    for i in range(1, 4):
        print("\nEnter details for contact", i)
        name    = input("Name: ")
        address = input("Address: ")
        mobile  = input("Mobile: ")
        email   = input("Email: ")
        writer.writerow([name, address, mobile, email])

print("\nData saved to address_book.csv successfully!")

# reading back and displaying the saved data
print("\n--- Address Book ---")
with open('address_book.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
import sqlite3

# -----------------------------------------------
# STEP 1 - Create Database
# -----------------------------------------------
conn = sqlite3.connect("school.db")
print("Database created: school.db")


# -----------------------------------------------
# STEP 2 - Create 3 Tables
# -----------------------------------------------

# table 1 - student (same as notes)
conn.execute('''
    CREATE TABLE IF NOT EXISTS student(
        st_id    INTEGER PRIMARY KEY AUTOINCREMENT,
        st_nm    VARCHAR(50),
        st_class VARCHAR(20),
        st_email VARCHAR(80)
    )
''')

# table 2 - teacher
conn.execute('''
    CREATE TABLE IF NOT EXISTS teacher(
        tc_id      INTEGER PRIMARY KEY AUTOINCREMENT,
        tc_nm      VARCHAR(50),
        tc_subject VARCHAR(50)
    )
''')

# table 3 - fees
conn.execute('''
    CREATE TABLE IF NOT EXISTS fees(
        fee_id  INTEGER PRIMARY KEY AUTOINCREMENT,
        st_id   INTEGER,
        amount  INTEGER,
        status  VARCHAR(20)
    )
''')

conn.close()
print("3 Tables created successfully!")


# -----------------------------------------------
# STEP 3 - Insert Records
# -----------------------------------------------
conn = sqlite3.connect("school.db")

# inserting into student table
conn.execute("INSERT INTO student(st_nm, st_class, st_email) VALUES('Rahul', '10A', 'rahul@gmail.com')")
conn.execute("INSERT INTO student(st_nm, st_class, st_email) VALUES('Pooja', '10B', 'pooja@gmail.com')")
conn.execute("INSERT INTO student(st_nm, st_class, st_email) VALUES('Arjun', '11A', 'arjun@gmail.com')")
conn.execute("INSERT INTO student(st_nm, st_class, st_email) VALUES('Sneha', '10A', 'sneha@gmail.com')")

# inserting into teacher table
conn.execute("INSERT INTO teacher(tc_nm, tc_subject) VALUES('Mr. Sharma', 'Maths')")
conn.execute("INSERT INTO teacher(tc_nm, tc_subject) VALUES('Ms. Verma', 'Science')")
conn.execute("INSERT INTO teacher(tc_nm, tc_subject) VALUES('Mr. Khan', 'English')")

# inserting into fees table
conn.execute("INSERT INTO fees(st_id, amount, status) VALUES(1, 5000, 'Paid')")
conn.execute("INSERT INTO fees(st_id, amount, status) VALUES(2, 5000, 'Pending')")
conn.execute("INSERT INTO fees(st_id, amount, status) VALUES(3, 5000, 'Paid')")

conn.commit()
conn.close()
print("Records inserted!")


# -----------------------------------------------
# STEP 4 - Different SELECT Operations
# -----------------------------------------------
conn = sqlite3.connect("school.db")

# select all students
print("\n--- All Students ---")
data = conn.execute("SELECT * FROM student")
for m in data:
    print(m)

# select with ORDER BY and LIMIT (from notes)
print("\n--- Top 2 Students sorted by name DESC ---")
data = conn.execute("SELECT * FROM student ORDER BY st_nm DESC LIMIT 2")
for m in data:
    print(m)

# filter/search by name (from notes)
print("\n--- Search Student by Name ---")
nm = input("Enter student name to search: ")
data = conn.execute("SELECT * FROM student WHERE st_nm = '" + nm + "'")
for m in data:
    print(m)

# select distinct classes (from notes)
print("\n--- Distinct Classes ---")
data = conn.execute("SELECT DISTINCT st_class FROM student")
for m in data:
    print(m)

# select all teachers
print("\n--- All Teachers ---")
data = conn.execute("SELECT * FROM teacher")
for m in data:
    print(m)

conn.close()


# -----------------------------------------------
# STEP 5 - Update Data
# -----------------------------------------------
conn = sqlite3.connect("school.db")

print("\n--- Before Update ---")
data = conn.execute("SELECT * FROM student ORDER BY st_nm DESC")
for m in data:
    print(m)

# update student name where st_id = 1 (same pattern as notes)
conn.execute("UPDATE student SET st_nm = 'Rahul Kumar' WHERE st_id = 1")
conn.commit()

print("\n--- After Update ---")
data = conn.execute("SELECT * FROM student ORDER BY st_nm DESC")
for m in data:
    print(m)

conn.close()


# -----------------------------------------------
# STEP 6 - Delete Data
# -----------------------------------------------
conn = sqlite3.connect("school.db")

print("\n--- Before Delete ---")
data = conn.execute("SELECT * FROM student ORDER BY st_nm DESC")
for m in data:
    print(m)

id = input("\nEnter student id to delete: ")
conn.execute("DELETE FROM student WHERE st_id = " + id)
conn.commit()

print("\n--- After Delete ---")
data = conn.execute("SELECT * FROM student ORDER BY st_nm DESC")
for m in data:
    print(m)

conn.close()
```
